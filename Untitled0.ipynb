**DEFINITION OF ALL OPERATOR4S IN PYTHON:-**

**(1) ARITHMATIC OPERATORS**:-
are used to perform mathematical operations like addition, subtraction, multiplication, etc.


Screenshot (30).png

```python
## EXAMPLE OF ARITHMATIC OPERATORS

a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.333...
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 1000
```

**2 COMPARISION OPERATORS:-** These operators are used to compare two values — the result is always either True or False.

Screenshot (31).png

```python
## EXAMPLES OF COMPARISION OPERATORS

a = 10
b = 5

print(a == b)   # False
print(a != b)   # True
print(a > b)    # True
print(a < b)    # False
print(a >= b)   # True
print(a <= b)   # False
```

**3 LOGICAL OPERATORS:-** These operators are used to combine conditional statements — they return either True or False depending on the logic.

Screenshot (32).png

```python
## EXAMPLES OF LOGICAL OPERATORS

a = 10
b = 5

print(a > 5 and b < 10)   # True  (both conditions are True)
print(a > 15 or b < 10)   # True  (one condition is True)
print(not(a == 10))       # False (because a == 10 is True, but 'not' reverses it)
```

**4 BITWISE OPERATORS:-** these are a little different from logical ones.

While logical operators (and, or, not) work with boolean values (True/False),

bitwise operators work with binary bits (0s and 1s).

Screenshot (33).png

```python
## EXAMPLE OF BITWISE OPERATORS

a = 5   # binary: 0101
b = 3   # binary: 0011

print(a & b)   # 1  (0001)
print(a | b)   # 7  (0111)
print(a ^ b)   # 6  (0110)
print(~a)      # -6
print(a << 1)  # 10 (1010)
print(a >> 1)  # 2  (0010)
```

***5 ASSIGNMENT OPERATORS:-*** Assignment operators are used to assign values to variables — and sometimes perform an operation and assignment together.

Screenshot (34).png

```python
## EXAMPLES OF ASSIGNMENT OPERATORS

x = 10
x += 5   # x = 15
x -= 3   # x = 12
x *= 2   # x = 24
x //= 5  # x = 4
x **= 3  # x = 64
print(x)
```

**6 MEMBERSHIP OPERATORS:-** These operators are used to check whether a value exists (or not) in a sequence — such as a string, list, tuple, or dictionary.

Screenshot (35).png

```python
## MEMBERSHIP OPERATORS

# Example with strings
text = "hello world"
print('h' in text)        # True
print('z' not in text)    # True

# Example with lists
fruits = ['apple', 'banana', 'mango']
print('apple' in fruits)       # True
print('grape' not in fruits)   # True

# Example with dictionary keys
student = {'name': 'Alice', 'age': 20}
print('name' in student)       # True
print('grade' not in student)  # True
```

**7 IDENTITY OPERATORS:-** These operators are used to compare the memory locations of two objects — in other words, they check whether both variables refer to the same object in memory (not just equal values).

Screenshot (36).png

```python
## EXAMPLES OF IDENTITY OPERATORS

x = [1, 2, 3]
y = x         # y refers to the same list as x
z = [1, 2, 3] # z is a different list with the same values

print(x is y)       # True  (same object)
print(x is z)       # False (different objects)
print(x == z)       # True  (same values)
print(x is not z)   # True  (different memory)
```

**DATA TYPE CONVERSION**:- Data type conversion means changing one type of value into another.

for example, changing a number into a string or a string into a float.

There are two types of conversions in Python:

**1️⃣ Implicit Conversion/type coercion. (Automatic):-**

Python automatically converts one data type to another when needed —
usually from smaller → larger data type (to avoid data loss).

**2️⃣ Explicit Conversion (Manual / Type Casting)**:-

 You manually change the data type using Python’s built-in functions like:
int(), float(), str(), bool(), etc.

data type conversion.png

```python
#  {IMPLICIT CONVERSION EXAMPLE}


a = 10
b = 3.5
result = a + b
print(result)          # 13.5
print(type(result))    # <class 'float'>
```

**line:1**

a is a variable.

It stores the value 10.

Since 10 has no decimal point, Python treats it as an integer (int).

**line:2**

b is another variable.

It stores the value 3.5.

Since 3.5 has a decimal, Python treats it as a floating-point number (float).

**line:3**

Here, you are adding an integer (a) and a float (b).

Python automatically performs implicit conversion:

It converts a (10) from int → float (10.0)

Then adds 10.0 + 3.5 = 13.5

This is called implicit type conversion or type coercion.

Now, result stores 13.5, which is a float.

**line:4**

Prints the value of result.✅ Output:

**line:5**

type() tells us the data type of result.
✅ Output:

```python
# EXPLICIT CONVERSION EXAMPLE

x = 12.8
y = int(x)
print(y)        # 12
print(type(y))  # <class 'int'>
```

**line:1**

Here, you created a variable named x.

You stored the value 12.8 in it.

Since 12.8 has a decimal point, Python treats it as a floating-point number (float).

**line:2**

Here, you are converting the float value (12.8) into an integer (int) using the built-in function int().

The int() function removes the decimal part — it does not round, it just cuts off everything after the decimal point.

So:
👉 int(12.8) → 12

Now y stores the value 12, and its data type is integer.

**line:3**

This prints the value stored in y.
✅ Output:

**line:4**

The type() function shows what kind of data y is holding.
✅ Output:

```python
#Implicit - Automatically
num1 = 12 #int
num2 = 4.3 #float

print('Number 1', num1,type(num1), 'Number', num2, type(num2))

num3 = num1 + num2

print('Number', num3, type(num3))
```

print('Number 1', num1, type(num1), 'Number', num2, type(num2))

Python automatically converts num1 → float before addition (this is implicit type conversion)

Then it adds: 12.0 + 4.3 = 16.3

So now:

num3 = 16.3

type(num3) = float

**Step 4: Printing result and type**

print('Number', num3, type(num3))

✅ Output

Number 16.3 <class 'float'>

Shows the value of num3 and its data type.




explicit datatype.png

```python
#Explicit

num1 = 12 #int
num2 = '4.3 '#str

print('Number 1', num1,type(num1), 'Number', num2, type(num2))

num3 = num1 + float(num2)

print('Number', num3, type(num3))
```

**UDERSTANDING ABOVE CODE**

**Step 1: Variable Assignment**

num1 = 12
num2 = '4.3 '

num1 → stores 12, which is an integer (int)

num2 → stores '4.3 ' (a string, note the quotes)

Anything inside quotes (' ' or " ") is considered a string (str) in Python

**Step 2: Print values and types**

print('Number 1', num1, type(num1), 'Number', num2, type(num2))

type(variable) shows the data type of each variable
✅ Output:

Number 1 12 <class 'int'> Number 4.3  <class 'str'>

**Step 3: Convert string to float and add**

num3 = num1 + float(num2)

float(num2) → converts the string '4.3 ' into a floating-point number 4.3

Then Python adds num1 + float(num2) → 12 + 4.3 = 16.3

Now num3 stores 16.3, which is a float (float)

This is called explicit type conversion (or type casting) because you manually convert the string to a float.

**Step 4: Print result and type**

print('Number', num3, type(num3))

✅ Output:

Number 16.3 <class 'float'>

Shows the value of num3

Confirms the data type is float

**Summary Table**

Variable	Value	Data Type	Notes

num1	12	int	Integer number

num2	'4.3 '	str	String (needs conversion to use as number)

num3	16.3	float	Result of addition (int + converted float)
